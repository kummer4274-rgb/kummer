# if, else

조건에 따라 실행을 달리하기 위한 키워드로 if, elif, else 가 있습니다.

````python
ret = 0.1

if ret > 0.0:
    print("상승")
    print("▲")
else:
    print("하락")
    print("▼")
    
print('같은 레벨의 들여쓰기로로 블럭을 구분합니다')
````

## if, elif, else 

````python
# True/False를 상승 여부로 가정

stock01 = False 
stock02 = True  

if stock01:
    print("종목01 선택")
elif stock02:
    print("종목02 선택")
else:
    print("아무것도 선택하지 않음")
````

````python
stock01 = True 
stock02 = True  

if stock01 == True:
    if stock02 == True:
        print("종목01 선택, 종목02 선택")
````

## for 와 range

```python
range(시작값, 종료값, 증가값)
```

```python
range(10) # 0~9 (0에서 시작, 종료값 10은 미포함)
range(1, 10) # 1~9
range(10, 20, 2) # 10~19, 2씩 건너뛰어서
```

````python
for i in range(10): # 0~9 (0에서 시작, 종료값 10은 미포함)
    print(i, end=', ')
````

````python
for i in range(1, 10): # 1~9
    print(i, end=', ')
````

````python
for i in range(10, 20, 2): # 10~19, 2씩 건너뛰어서
    print(i, end=', ')
````

````python
for i in range(0, 100+1, 20): # 0~100, 20씩 건너뛰어서
    print(i, end=', ')
````

````python
[x for x in range(-5, 6, 2)]
````

다른 예시도 몇가지 살펴보죠.

````python
# -5 ~ +5
for x in range(-5, 5+1):
    print(x, end=', ' )
````

````python
# 0 ~ 10까지 모든 짝수
for x in range(0, 10+1, 2):
    print(x, end=', ' )
````

list() 사용하여 range를 리스트로 변환할 수 있다는 점도 참고로 보아두십시오.

````python
print( list(range(10)) )
print( list(range(1, 10)) )
print( list(range(10, 20, 2)) )
````

# break, continue

* break: 반복문을 종료
* continue: 반복문의 시작으로 이동

````python
for ch in "인생은 짧아요":
    if ch == "은":
        break
    print(ch)

print(">> 반복문 종료")
````

````python
for ch in "인생은 짧아요":
    if ch == "은":
        continue
    print(ch)

print(">> 반복문 종료")
````

# 리스트 컴프리헨션

[`표현식` for `항목` in `순회_가능_객체`]

````python
#  0~9까지 각 수의 제곱

[x ** 2 for x in range(0, 10)]
````

[`표현식` for `항목` in `순회_가능_객체` if `조건식`]

````python
even_squares = [x ** 2 for x in range(1, 10) if x % 2 == 0]
even_squares
````
