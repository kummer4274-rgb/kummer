## 국내증시 주요지표

```python
import requests
from bs4 import BeautifulSoup

url = 'https://finance.naver.com/sise/'# (국내증시)
r = requests.get(url)
soup = BeautifulSoup(r.text)

span = soup.find(id='KOSPI_now')
span.text
```

```python
float(span.text.replace(',', ''))
```

```python
# 타입 변환 (str >> float)
span = soup.find(id='KOSDAQ_now')
float(span.text.replace(',', ''))
```

## read_html()로 테이블 가져오기

<img src='https://i.imgur.com/9dFTKVK.png'>

```python
import pandas as pd

url = 'https://finance.naver.com/marketindex/'
df_list = pd.read_html(url, encoding='euc-kr')
df = df_list[0] # 첫번째 테이블
df
```

## 엑셀로 저장하기

```python
df.to_excel('국내금리.xlsx')
```
