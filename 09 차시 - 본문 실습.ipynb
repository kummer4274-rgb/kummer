## 예외 처리 try ... except

````python
stocks = ['삼성전자', 'SK하이닉스', '현대차']
````

````python
print(stocks[4])
print("이후 실행 내용")   
````

````python
try:
    print(stocks[4])
except:
    print('예외가 발생했습니다.')
    
print("이후 실행 내용")   
````

````python
try:
    print(stocks[4])
except IndexError as e:
    print('예외가 발생했습니다.')
    print(e)
    
print("이후 실행 내용")   
````

````python
x = [10, 20, 30]

try:
    print(x[1] / 0)  # ZeroDivisionError
    print(x[4])      # IndexError
except Exception as e:
    # 그밖에 모든 예외 처리하기
    print('예외가 발생했습니다')
    print(e)
    
print("이후 실행 내용")    
````

````python
x = [10, 20, 30]

try:
    print(x[1] / 0)  # ZeroDivisionError
    print(x[4])      # IndexError
except IndexError as e:
    # 내가 예외 처리하기
    print('인덱스의 범위를 확인하세요')
    print(e)
except Exception as e:
    # 그밖에 모든 예외 처리하기
    print('예외가 발생했습니다')
    print(e)

print("이후 실행 내용")     
````

## 파일 입출력

````python
data = "6, 12,  3,  2, 13,  2, 10, 15, 19,  7"
print(data)
````

````python
f = open("output.txt", "w")
f.write(data)
f.close()
````

````python
f = open("output.txt", "r")
contents = f.read()
f.close()

print(contents)
````

````python
# 쓰기
data = "6, 12,  3,  2, 13,  2, 10, 15, 19,  7"

with open("output.txt", "w") as f:
    f.write(data)
````

````python
# 읽기
with open("output.txt", "r") as f:
    contents = f.read()
    
print(contents)
````
