1) 3의 배수를 다음과 같이 출력하는 코드를 작성해 보고, 

```
3, 6, 9, 12, 15, 18, 21, 24, 27
```

````python
[x for x in range(1,10)]
````

````python
[x*3 for x in range(1,10)]
````

````python
[x for x in range(1, 50) if x % 3 == 0]
````

2) 구구단을 다음과 같이 출력하는 코드를 작성해 보세요.
```
2 x 1 = 2
2 x 2 = 4
2 x 3 = 6
2 x 4 = 8
2 x 5 = 10
2 x 6 = 12
2 x 7 = 14
2 x 8 = 16
2 x 9 = 18
… 이하 생략 …
```

````python
for x in range(1,10):
    print('2 x {} = {}'.format(x, 2 * x))
````

````python
for g in range(2,10):
    for x in range(1,10):
        print('{} x {} = {}'.format(g, x, g * x))
    print()
````

````python
## 구구단 출력
````

````python
for i in range(2, 10):
  for j in range(1, 10):
    print(i, 'x', j, '=', i * j)
  print()
````
