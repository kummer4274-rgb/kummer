## str.format() 포맷팅 (파이썬 3)

```python
# str.format()  포맷팅 (Python3 스타일)

# 변수에 값을 할당
a, b, c = 3.1415, 2.8, 0.015

# 순서대로 할당
print("x = {:.2f}, y = {}, z = {}".format(a, b, c))

# 번호로 순서를 지정
print("x = {0:.2f}, y = {2}, z = {1}".format(a, b, c))

# 변수명을 사용하여 지정
print("x = {x:.2f}, z = {z}, y = {y}".format(x=a, y=b, z=c))
```

##  f-string (파이썬 3.6 이상)

## 문자열 인덱싱(indexing)과 슬라이싱(slicing)

#### 인덱싱(indexing)
`s[index]`


#### 슬라이싱(slicing)
`s[start:end]`


```python
s = "Hello, Python!"

# 인덱싱
print(s[0]) # 첫번째 문자
print(s[4]) # 5번째 문자

# 슬라이싱
print(s[:]) # 처음 부터 끝까지
print(s[0:5]) # 처음 부터 인덱스 4까지
print(s[7:9]) # 인덱스 7부터, 인덱스 8까지
print(s[-1]) # 마지막 문자
print(s[-5:-1]) # 마지막에서 5번째 부터 마지막 문자 직전까지
```

## 문자열 자르기와 합치기 그리고 바꾸기

* `s.split(지정문자)`: 문자열 s를 지정문자(기본값은 공백 문자)를 기준으로 잘라 그 결과를 리스트로 반환
* `s.join(리스트)`: 리스트들을 문자열 s를 구분자로 하여 붙여 하나의 문자열을 생성
* `s.replace(현재문자열, 대체문자열)`: 문자열 s에 있는 '현재문자열'을 '대체문자열'로 대체

```python
# 문자열 분리(split)

s = "화학,출판,전기제품,제약,은행"
s.split(',')
```

문자열을 콤마(,)를 기준으로 분리 합니다.

```python
# 문자열 합치기 (join)

sectors = ['화학', '출판', '전기제품', '제약', '은행']
"|".join(sectors)
```

```python
# 문자열 대체 (replace)
s = "Hello, Python!"
w = s.replace('Python', 'World')
w
```

```python
s = '8,832,934원'
price = int(s.replace(',', '').replace('원', ''))

print(price)
```
