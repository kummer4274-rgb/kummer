## 슬라이싱(slicing)

lst[`start`:`end`] <br>
lst[`start`:`end`:`step`]

````python
stocks = ['삼성전자', 'SK하이닉스', '현대차', '셀트리온', 'LG화학', 'NAVER']

print(stocks[1:4]) # 1~3
print(stocks[:3]) # 앞부분 3개 (0~2)
````

````python
print(stocks[:3])  # 0~2 (처음 3개)
print(stocks[4:])  # 4~끝까지
print(stocks[2:3]) # 2 (1개)
print(stocks[4:])  # 4~끝까지
print(stocks[:])   # 처음부터~끝까지
print(stocks[-1])  # 마지막 요소
print(stocks[:-1]) # 마지막를 제외하고 전체
````

````python
lst = ['a', 'b', 'c', 'd', 'e']

lst[:]       # ['a', 'b', 'c', 'd', 'e']
lst[3:]     # ['d', 'e']
lst[:-1]    # ['a', 'b', 'c', 'd']
lst[1:3]   # ['b', 'c']
lst[1:-1]  # ['b', 'c', 'd']
lst[-3:-1] # ['c', 'd']
lst[::-2]   # ['e', 'c', 'a']
lst[::2]    # ['a', 'c', 'e']
lst[::-1]   # ['e', 'd', 'c', 'b', 'a']
````

## 리스트 덧셈 (합치기)

## 딕셔너리와 반복문 

````python
corps = {}
corps['AAPL'] = 'AAPL Apple Inc.'
corps['GOOG'] = 'Alphabet Inc.'
corps['GOOGL'] = 'Alphabet Inc.'
corps['MSFT'] = 'Microsoft Corporation'
corps['AMZN'] = 'Amazon.com, Inc.'
````

````python
corps
````

딕셔너리를 for..in 에 사용하면 키들을 차례로 추출합니다.

````python
for key in corps:
    print(key, ':', corps[key])
````

딕셔너리의 items() 메소르를 사용하면 key,value 전체를 리스트로 얻을 수 있습니다.

````python
corps.items()
````

따라서 다음과 같이 for문을 순회하면 key,value를 차례로 가져와 처리할 수 있습니다.

````python
for key, val in corps.items():
    print (key, ':', val)
````
