* 특정한 날짜부터 현재까지 경과 시간 구하기
* 경과 시간을 “n년 m일” 형식으로 표시하기
* 특정한 날짜의 요일(예 : “금요일”) 구하기


```python
# 답안
from datetime import datetime

the_day = datetime(1998, 2, 17)
delta = datetime.now() - the_day
delta
```

```python
# 답안
from datetime import datetime

the_day = datetime(1998, 2, 17)
delta = datetime.now() - the_day

"{}년 {}일".format(delta.days // 365, delta.days % 365)
```
