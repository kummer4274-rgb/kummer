# 10 차시 - 연습문제 해설

통화에는 본원통화(base money)외에 M1(협의의 통화)와 M2(광의통화)가 있습니다.

* M1(협의통화)	= 현금통화 + 요구불예금 + 수시입출식 저축성예금
* M2(광의통화)	= M1 + 정기예적금 + 시장형 상품 + 실적배당형 상품 + 금융채

투자 혹은 소비심리는 가늠하기 위해 통화량의 `M1/M2 비율`을 유동성을 측정하는 하나의 지표로 살펴보기도 합니다. `M1/M2 비율`이 클수록 현금유성동이 크다고 할 수 있습니다. 다음 내용을 구현해 봅니다.

1. FRED로 부터 한국의 M1, M2 통화량 데이터를 가져옵니다
1. M1, M2의 데이터를 가지는 DataFrame 만듧니다.
1. `M1/M2 비율`을 구해서 `liquidity` (유동성)컬럼을 생성합니다.
1. liquidity (유동성)컬럼을 차트로 그립니다.
1. 시기에 따른 `유동성` 컬럼의 변화로 우리나라 유동성이 높았을때와 낮았을때 어떤 배경이 있었는지 토론합니다.

힌트:

FRED에서 한국의 M1, M2 데이터 페이지는 다음과 같습니다. (월별 데이터이며, 1970년 ~ 2017년 5월 까지 데이터가 제공됩니다)
* https://fred.stlouisfed.org/series/MYAGM1KRM189S
* https://fred.stlouisfed.org/series/MYAGM2KRM189S

pandas_datareader의 DataReader()함수를 사용해 fred로 부터 데이터를 읽어 올 수 있습니다.



```python
import pandas_datareader as pdr

df = pdr.DataReader(['MYAGM1KRM189S', 'MYAGM2KRM189S'], 'fred', '1990')
df
```

```python
%matplotlib inline
```

```python
df.plot(figsize=(10,6))
```

```python
df['liquidity'] = df['MYAGM1KRM189S'] / df['MYAGM2KRM189S']
df['liquidity']
```

```python
df['liquidity'].plot(figsize=(10,6))
```
