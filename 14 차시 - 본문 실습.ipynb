## Matplotlib 차트 그리기

```python
# 주피터 노트북 페이지 안에 matplotlib 차트를 표현하는 옵션 
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime

# 전역 값 설정
plt.rcParams['figure.figsize'] = (14,4)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'b'
plt.rcParams['axes.grid'] = True
```

## 간단한 그리기 1

```python
x = np.array([0, 1, 2, 3, 4, 5, 6, 7 ])
y = x ** 2

plt.plot(x, y, 'r')
plt.show()
```

## 간단한 그리기 2

```python
data1 = np.random.randint(-50,50,100).cumsum()
data2 = np.random.randint(-50,50,100).cumsum()

plt.plot(data1)
plt.plot(data2)
```

## 타이틀, 레이블, 범례

타이틀 title, 레이블 label, 범례 legend

```python
fig, ax = plt.subplots(figsize=(16,4))

x = np.arange(100)
y1 = np.random.randint(-50,50,100).cumsum()
y2 = np.random.randint(-50,50,100).cumsum()
y3 = np.random.randint(-50,50,100).cumsum()

ax.plot(x, y1, 'r')
ax.plot(x, y2, 'g')
ax.plot(x, y3, 'b')

ax.set_title("title") #  타이틀
ax.set_xlabel("x") # 레이블
ax.set_ylabel("y") 
ax.legend(["Red", "Green", "Blue"]) # 범례
```

 ## 색상

```python
# 하나의 색상(color)는 RGB 값을 16진수로 표현

x = np.arange(0, 5, 0.5)

plt.plot(x, x**2.0, color='#FF0000') # 'red' = 'r' = '#FF0000'
plt.plot(x, x**2.5, color='green')   # 'green' = 'g', ''#00FF00'
plt.plot(x, x**3.0, color='b') # B
```

## Matplotlib 여러가지 차트 그리기


## Stack Plots

```python
ar1 = [64,  5, 62, 65, 35, 24,  9, 18, 50]
ar2 = [17,  5, 14,  2, 12,  6, 19, 11, 18]
ar3 = [63, 71, 66, 70, 67, 78, 65, 65, 65]
inx = [1,  2,  3,  4,  5,  6,  7,  8,  9]

plt.stackplot(inx, ar1, ar2, ar3, colors= ['r','g','b'])
plt.title('Stack Plot')
```

# Pie Charts

```python
labels = 'S1', 'S2', 'S3'
data = [56, 66, 24]
colors = ['c', 'm', 'y']

plt.pie(data, labels=labels, colors=colors,
        startangle=90,
        explode = (0, 0.1, 0),
        autopct = '%1.2f%%')

plt.title('Pie Chart Example')
plt.show()
```

## Bar chart

```python
x = np.linspace(0,5,6)
plt.bar(x, x**2, align="center", width=0.5, alpha=0.5)
```

## Histogram

```python
n = 5 + np.random.randn(1000)

plt.hist(n, bins=20)
plt.title("Histogram")
plt.show()

plt.hist(n, cumulative=True, bins=20)
plt.title("Cumulative Histogram")
plt.show()
```

## fill_between and alpha

```python
plt.fill_between(x, x**2, x**3, color="green", alpha=0.5)
```

```python
ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
```

## figure(도형)와 axes(축)

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.9]) # main
ax2 = fig.add_axes([0.2, 0.5, 0.3, 0.3]) # inner

x = np.arange(100)
y = np.random.randn(100).cumsum()

ax1.plot(x, y, 'r') # main chart
ax1.set_xlabel('x'); ax1.set_ylabel('y')

x = np.arange(100)
y = np.random.randn(100).cumsum()

ax2.plot(x, y, 'g') # inner chart
ax2.set_title('inner chart')
```

## 예제 데이터셋 - 팁 데이터 (회귀 문제)


```python
import seaborn as sns

tips = sns.load_dataset("tips")

print(tips.shape)
tips.sample(10) # 임의 10개 데이터
```

## relplot (relation plot)

```python
sns.relplot(x="total_bill", y="tip", data=tips);
```

hue를 지정하여 종류를 구분

```python
ax = sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

style 을 지정하여 더 강조

```python
ax = sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker", data=tips)
```

연속적인 값일 때 hue를 지정하면 단계별로 색이 달라집니다

```python
ax = sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
```

size를 지정하면 크기를 다르게 표현합니다

```python
ax = sns.relplot(x="total_bill", y="tip", hue="size", size="size", data=tips)
```
