## 날짜시간 포맷팅

```python
from datetime import datetime

dt = datetime(2019, 3, 10, 11, 30)
```

```python
dt.isoformat()
```

```python
print ("%d-%02d-%02d" % (dt.year, dt.month, dt.day))
print ("%02d:%02d:%02d" % (dt.hour, dt.minute, dt.second))
print ("%d년 %d월 %d일 %s요일" % (dt.year, dt.month, dt.day, '월화수목금토일'[dt.weekday()]))
```

```python
print ("{:d}-{:02d}-{:02d}".format(dt.year, dt.month, dt.day))
print ("{:02d}:{:02d}:{:02d}".format(dt.hour, dt.minute, dt.second))
print ("{}년 {}월 {}일 {}요일".format(dt.year, dt.month, dt.day, '월화수목금토일'[dt.weekday()]))
```

## 날짜시간을 문자열로, 문자열을  날짜시간으로

* datetime.strftime(): datetime → 문자열  
* datetime.strptime(): 문자열 → datetime

```python
from datetime import datetime
dt = datetime(2019, 3, 10, 11, 30)
```

```python
# 시간을 문자열로  strftime()

print( dt.strftime("%Y-%m-%d") )
print( dt.strftime("%H:%M:%S") )
print( dt.strftime("%Y-%m-%d %H:%M:%S") )
print( dt.strftime("%Y년 %m월 %d일 %H시 %M분") )
```

```python
# 문자열을 시간으로 strptime()

s = "2019-03-10 11:30:00"
dt = datetime.strptime(s, "%Y-%m-%d %H:%M:%S")
print(dt)

s = "2019년 03월 10일 11시 30분"
dt = datetime.strptime(s, "%Y년 %m월 %d일 %H시 %M분")
print(dt)
```

## 다양한 데이터 다루기 - 컬렉션(collection)

```python
from collections import Counter

stocks = ['삼성전자', 'SK하이닉스', '현대차', '셀트리온', '셀트리온', 'LG화학', 
          'LG화학', 'LG화학', '현대차', '삼성전자', '셀트리온', '셀트리온']

cntr = Counter(stocks)
cntr.most_common()
```
