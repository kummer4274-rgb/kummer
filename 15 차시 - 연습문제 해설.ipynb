```python
import glob
import pandas as pd

fn_list = glob.glob('*_AD_2019.xlsx')
fn_list
```

```python
fn_list[0]
```

```python
df = pd.read_excel(fn_list[0], index_col='date', parse_dates=True)
df
```

```python
df['total']
```

```python
for fn in fn_list:
    print(fn)
```

```python
merged = pd.DataFrame()

for fn in fn_list:
    df = pd.read_excel(fn, index_col='date', parse_dates=True)
    corp = fn.split('_')[0]
    print(corp)
    merged[corp] = df['total']
```

```python
merged
```

```python
merged.to_excel('merged.xlsx')
```
