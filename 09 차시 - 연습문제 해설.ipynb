## 연습문제
다음과 같은 데이터가 저장된 파일(data.txt)이 있습니다. 이 데이터 파일을 읽어 0을 제외한 값들의 평균을 구하시오.

1. 파일 전체를 읽습니다. 파일이 없으면 파일이 없다는 메시지를 출력하도록 예외처리를 합니다.
1. 컴마(,)를 기준으로 텍스트를 분리하여 리스트를 만듭니다.
1. 리스트의 모든 값을 수치값을 바꿉니다.
1. 리스트에서 0을 모두 제거합니다. 
1. 전체 값의 평균을 계산합니다. 


```python
%%writefile data.csv
74,70,65,0,19,42,94,67,0,76,75,0,65,0,58,90
```

## 1) 파일 전체를 읽습니다. 

파일이 없으면 파일이 없다는 메시지를 출력하도록 예외처리를 합니다.

없는 파일을 오픈하는 경우 FileNotFoundError 예외가 발생합니다

```python
with open('data.csv', 'r') as f:
    text = f.read()
```

```python
try:
    with open('data.txt', 'r') as f:
        text = f.read()
except FileNotFoundError as e:
    print('파일이 존재하지 않습니다', e)
```

데이터 파일을 생성합니다.

```python
%%writefile data.csv
74,70,65,0,19,42,94,67,0,76,75,0,65,0,58,90
```

```python
try:
    with open('data.csv', 'rt') as f:
        text = f.read()
except FileNotFoundError as e:
    print('파일이 존재하지 않습니다', e)
```

```python
text
```

## 2) 컴마를 기준으로 텍스트를 분리하여 리스트를 만듭니다. 

특정 문자열을 기준으로 분리하여 리스트를 만들려면 split 함수를 사용합니다




```python
data = text.split(',')
print(data)
```

## 3) 리스트의 모든 값을 수치값을 바꿉니다.
int(x)는 x를 정수로 변환합니다. 리스트 컴프리헨션을 활용하면 더 좋습니다



```python
values = [int(x) for x in data]
values
```

## 4) 리스트에서 0을 모두 제거합니다.

0이 있는 조건이면 리스트에서 '0'을 제거합니다. in 연산자는 요소가 있는지 확인하는데 사용합니다. while은 특정 조건인 동안 반복합니다

```python
while 0 in values:
    values.remove(0)

values
```

```python
# while 문을 쓰지 않고 다음과 같이 리스트 컴프리헨션의 if 문을 사용할수도 있습니다.

values = [x for x in values if x != 0]
values
```

## 5) 전체 값의 평균을 계산합니다. 
sum(), len()과 같은 내장 함수를 활용하면 간편하게 계산할 수 있습니다

```python
sum(values) / len(values)
```
