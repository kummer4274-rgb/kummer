````python
def get_sign(x): 
    if x > 0: 
        return '양수' 
    elif x < 0: 
        return '음수' 
    else: 
        return '영' 

for x in [-1, 0, 1]: 
    print (get_sign(x))
````

* all(), any() - 다수의 True/False 조건을 테스트하는 함수들

````python
print(len("파이썬"))      # str은 문자열의 길이(=문자의 개수)
print(len(["파이썬", "쓰세요"]))
print(len([10, 20, 30, 40, 50]))
print(len([10, 'Hello', (30, 40, 50)]))
````

## min(), max(), sum() - 계산에 사용하는 함수들

* min(`이터러블`) - `이터러블`에서 가장 작은 최솟값을 반환 
* max(`이터러블`) - `이터러블`에서 가장 큰 최댓값을 반환
* sum(`이터러블`) - `이터러블`의 합을 반환
* round(`숫자값`, `자릿수`) - `숫자값`을 소수점이하 `자릿수`만큼 정밀도로 반올림한 값을 반환 (`자릿수`를 생략하면 정수값을 반환)

````python
data = [10, 20, 30, 50, 110, 2, 90, 30, 8]

print('데이터 합계:', sum(data))
print('데이터 최댓값:', max(data))
print('데이터 최솟값', min(data))
````

````python
len(data)
````

data의 평균을 구하고자 한다면,

````python
average = sum(data) / len(data)
average
````

````python
round(average, 2) # 소수점 2자리 반올림
````

## all(), any() - 다수의 True/False 조건을 테스트하는 함수

* all(`이터러블`) - `이터러블`의 모든 값이 참이면 True를 반환
* any(`이터러블`) - `이터러블`중에 하나라고 참이면 True를 반환

````python
print( all([True, True, True]) )
print( any([True, False, False]) )
````
