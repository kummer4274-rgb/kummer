## 코드적용 결과 검사
결측치(인덱스로 매칭되지 않은 항목) 찾기

```python
# 결측치(인덱스로 매칭되지 않은 항목) 찾기
fn_list = glob.glob('*코드적용.xlsx')
for fn in fn_list:
    payroll =  pd.read_excel(fn)
    missing = payroll[payroll['코드'].isnull()]
    if len(missing):
        print(fn, len(missing), '건 오류')
        print('-' * 50)
        print(missing)
    else:
        print(fn, 'OK')
```

이런 내용이 발견되면 은행코드 엑셀 파일(bank_codes.xlsx)에 은행명칭을 추가하고 변환을 다시 수행하면 됩니다.

은행코드 엑셀 파일(bank_codes.xlsx)에 'KDB산업은행(002)'과 '카카오(090)'을 추가하고 다시 수행합니다.

## 전체코드

```python
import pandas as pd
import glob

fn_bankcode = 'payroll/bank_codes.xlsx'
bank_codes = pd.read_excel(fn_bankcode, index_col='은행', dtype={'코드':str})

# 모든 엑셀 파일에 대해 처리
fn_list = glob.glob('payroll/corp_*.xlsx')
for fn in fn_list:
    # 은행코드 컬럼 추가
    payroll = pd.read_excel(fn, index_col='은행')
    payroll['코드'] = bank_codes['코드']
    to_fn = fn.replace('payroll/', '').replace('.xlsx', '_코드적용.xlsx')
    payroll.to_excel(to_fn)

    # 은행코드 적용 결과 검사
    payroll =  pd.read_excel(to_fn)
    missing = payroll[payroll['코드'].isnull()]
    if len(missing):
        print(fn, len(missing), '건 오류')
        print('-' * 50)
        print(missing)
    else:
        print(fn, '-->', to_fn, 'OK')
```
