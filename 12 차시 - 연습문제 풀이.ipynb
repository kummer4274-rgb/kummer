`종목코드`의 타입을 str로 지정하고 있습니다. 종목코드는 숫자로만 구성되어 있기 때문에 타입을 str로 지정하지 않으면 수치값으로 읽게 됩니다.

1. stock_factors 에 있는 있는 컬럼(PER,PBR,주당배당금,배당수익률) 데이터를 stock_fav 로 옮깁니다

1. 잘 저장되었는지 저장된 CSV를 확인합니다.


````python
import pandas as pd

stock_fav = pd.read_csv('https://bit.ly/3g2h1WL', dtype={'종목코드':str})
stock_fav
````

````python
stock_factors = pd.read_csv('https://bit.ly/3bB2kWZ', dtype={'종목코드':str})
stock_factors
````

## 인덱스 지정

````python
stock_fav.set_index('종목코드', inplace=True)
stock_factors.set_index('종목코드', inplace=True)
````

````python
# stock_fav = stock_fav.set_index('종목코드')
````

## 컬럼 복사

````python
stock_fav['PER'] = stock_factors['PER']
stock_fav['PBR'] = stock_factors['PBR']
stock_fav
````

````python
stock_fav[['주당배당금', '배당수익률']] = stock_factors[['주당배당금', '배당수익률']]
stock_fav
````

````python
stock_fav.to_csv('merged.csv')
````
